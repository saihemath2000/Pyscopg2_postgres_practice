```python
inspect_query = """
SELECT 
    level1_code AS code,
    level1_name AS name,
    1 AS level,
    jsonb_build_object(
        'sum_sales_index', SUM(COALESCE((metric_data->>'Sales Index')::float, 0)),
        'sum_assets_covered', SUM(COALESCE((metric_data->>'Assets Covered')::float, 0)),
        'sum_pm_count', SUM(COALESCE((metric_data->>'PM Count')::float, 0)),
        'sum_labor_hours', SUM(COALESCE((metric_data->>'Labor Hours')::float, 0)),
        'sum_travel_hours', SUM(COALESCE((metric_data->>'Travel Hours')::float, 0)),
        'sum_alignment_index', SUM(COALESCE((metric_data->>'Alignment Index')::float, 0)),
        'sum_machine_count', SUM(COALESCE((metric_data->>'Machine Count')::float, 0)),
        'sum_calls', SUM(COALESCE((metric_data->>'Calls')::float, 0))
    ) AS metric_data
FROM (
    SELECT 
        level1_code,
        level1_name,
        metric_data::jsonb AS metric_data
    FROM qa_mergetest5_position_geo_temp
    WHERE level1_code IS NOT NULL
) AS subquery
GROUP BY level1_code,level1_name;
"""
```

```python
# Connect to the database
try:
    conn = psycopg2.connect(**db_params)
    cursor = conn.cursor()
    
    # Execute the query
    cursor.execute(inspect_query)
    
    # Fetch all results
    results = cursor.fetchall()
    
    # Define column names
    columns = [desc[0] for desc in cursor.description]
    
    # Print results
    for row in results:
        row_dict = dict(zip(columns, row))
        print(json.dumps(row_dict, indent=2))
        
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Close the cursor and the connection
    cursor.close()
    conn.close()
```
